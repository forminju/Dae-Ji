문제 [1-1] - 8 점
범위 0과 50 사이의 소수(prime number)을 출력하는 코드를 작성하세요.

Hint: for 반복문을 사용해주세요.

(공백, 줄바꿈 상관 없이 출력 결과만 맞으면 정답 처리)

```python
list1=[]
for i in range(0,51):
  count =0

  for k in range(1,i+1):
       if i%k==0:
          count +=1
  if count <=2:
       print(i) 
       list1.append(i)
```

```python
print(list1)
```

```python
import random
```

```python
def rand_prime():
  print(random.choice(list1))
```

```python
rand_prime()
```
